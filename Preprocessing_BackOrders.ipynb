```python
# Checking for garbage values and storing in list
LL=[]
for i in range(0,len(df)):
  try:
    if (df['national_inv'][i]).isdigit()==False:
      LL.append(df['national_inv'][i])
    else:
      df['national_inv'][i]=int(df['national_inv'][i])
  except:
    continue
```

```python
# Removing garbage values using list
for i in range(0,len(df)):
  if df['national_inv'][i] in LL:
    df=df.drop(index=i)
```

```python
# Removing negative values from column
for i in df.index:
  if df['perf_6_month_avg'][i]<0:
    df=df.drop(index=i)
```
